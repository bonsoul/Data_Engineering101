```python
import json
import os
from pprint import pprint

# File paths
base_path = r'D:\Documents\Data Engineering\Data_Engineering101\data-engineering-essentials\035 Python Essentials for Data Engineers\05 Project 1 - File Format Converter\data\retail_db'
files = [
    'inpatient_2011 (1).json',
    'inpatient_2012 (1).json',
    'inpatient_2013 (1).json'
]

# Load all datasets into a list
datasets = []
for file in files:
    with open(os.path.join(base_path, file), 'r') as f:
        data = json.load(f)
        datasets.append(data)

# Unpack each dataset
data_2011, data_2012, data_2013 = datasets

# View column names for each file
columns_2011 = set(data_2011[0].keys())
columns_2012 = set(data_2012[0].keys())
columns_2013 = set(data_2013[0].keys())

print("2011 Columns:", columns_2011)
print("2012 Columns:", columns_2012)
print("2013 Columns:", columns_2013)

# Compare all three
common_columns = columns_2011 & columns_2012 & columns_2013
print("\n✅ Common Columns in All Years:", common_columns)

# Merge all records into a single dataset (optional)
combined_data = data_2011 + data_2012 + data_2013

# Optional: convert to pandas DataFrame
import pandas as pd
df = pd.DataFrame(combined_data)

# Show summary
print(f"\nCombined dataset has {len(df)} records and {len(df.columns)} columns.")
pprint(df.head(3).to_dict(orient="records"))
```

```python
#save as csv
output_path = os.path.join(base_path, 'inpatient_combined.csv')
df.to_csv(output_path, index=False)

print(f"\n Combined CSV to: {output_path}")
```
